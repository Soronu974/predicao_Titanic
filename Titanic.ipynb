# Tratamento dos dados


```python
#Excluindo colunas desnecessarias
df_treino = df_treino.drop(['Name','Ticket','Cabin'],axis=1)
df_teste = df_teste.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
df_treino.head(2)
```

```python
df_treino.Age.mean()
```

```python
#Preenchendo idades vazias com a média das idades no df
df_treino.loc[df_treino.Age.isnull(),'Age'] = df_treino.Age.mean()
df_teste.loc[df_teste.Age.isnull(),'Age'] = df_teste.Age.mean()
```

```python
df_treino.Embarked.value_counts()
```

```python
#Preenchendo valores vazios da coluna 'Embarked' com valor S
df_treino.loc[df_treino.Embarked.isnull(),'Embarked'] = df_treino.Embarked.mode()[0]
df_teste.loc[df_teste.Embarked.isnull(),'Embarked'] = df_teste.Embarked.mode()[0]
```

```python
df_teste.loc[df_teste.Fare.isnull(),'Fare'] = df_teste.Fare.mean()
```

#Pre-processamento

```python
#Transformando colunas de texto em numeros

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for columns in df_treino:
  if df_treino[columns].dtype == "object":
    df_treino[columns] = encoder.fit_transform(df_treino[columns])

df_treino.info()
```

```python
display(df_treino)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for columns in df_teste:
  if df_teste[columns].dtype == "object":
    df_teste[columns] = encoder.fit_transform(df_teste[columns])

df_teste.info()
```

```python
df_treino.info()
```

```python
X = df_treino.drop(['PassengerId','Survived'], axis=1)
y = df_treino.Survived
```

```python
X.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_val,y_train,y_val = train_test_split(X,y, train_size=0.3)
```

# Treinamento

```python
from sklearn.tree import DecisionTreeClassifier

arvore = DecisionTreeClassifier()
arvore = arvore.fit(X_train,y_train)
```

```python
y_pred_arvore = arvore.predict(X_val)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
knn = knn.fit(X_train,y_train)
```

```python

y_pred_knn = knn.predict(X_val)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr = lr.fit(X_train,y_train)

y_pred_lr = lr.predict(X_val)
```

#Avaliação de modelos

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_val,y_pred_arvore)
```

```python
# Convertendo valores do y_pred_knn para valores binários
y_pred_knn_binary = [1 if pred >= 0.5 else 0 for pred in y_pred_knn]

accuracy_score(y_val, y_pred_knn_binary)
```

```python
accuracy_score(y_val,y_pred_lr)
```
